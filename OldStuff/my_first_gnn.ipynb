```python
import numpy as np
import dgl
import networkx as nx
```

```python
src = [0,1,2,3]
dst = [3,2,3,1]

graph = dgl.graph((src,dst))
```

```python
print(graph)
```

```python
nx_g = graph.to_networkx().to_undirected()
pos = nx.kamada_kawai_layout(nx_g)
nx.draw(nx_g, pos, with_labels=True, node_color=[[.7, .7, .7]])
```

```python
nx_g_dir = graph.to_networkx().to_directed()
pos = nx.kamada_kawai_layout(nx_g_dir)
nx.draw(nx_g_dir, pos, with_labels=True, node_color=[[.7, .7, .7]])
```
